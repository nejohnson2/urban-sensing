### Temporary Expert
There is a dividing line that sensors run in letting people be a temporary expert.  A sensor can enable someone to collect data about a specific topic though this does not necessarily mean that they understand the topic.  For example, hydrology is a very complicated field and is needed to really understand the complexities of water quality.  Just taking a reading does not mean you're an expert.  On the other side of that coin, the sensor itself can enable people who do not understand the topic to actually go out an measure something useful.

### 'Minimal first pass on actually doing a sensor project'
1. What is the question you're trying to answer?  What is your expected outcome?
2. What are you going to measure to answer that question?  How does industry/government/experts try to answer that question?  Are there any standards for taking that measurement?
3. Choose your method of measuring
4. Know what the sensor is measuring.  What does it actually output?  How does it create that output.  Is it a phycial change in electriciy, a chemical change etc...?
5. Calibration - can the sensor measure the samet thing over and over?
6. Accuracy/Precision
7. Account for biases - before you collect data, plan how you're going to do it in order to avoid contamination, biases etc..
8. Measure Something - and make sure you provide an error bar!
9. Verify results 
10. Go back to number one and repeat
